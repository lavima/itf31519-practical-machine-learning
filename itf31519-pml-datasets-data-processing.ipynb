```python
# Load the UCI wine dataset into a Pandas dataframe

uci_wine_names = ['Target',
                  'Alcohol',
                  'Malic acid',
                  'Ash',
                  'Alcalinity of ash',
                  'Magnesium',
                  'Total phenols',
                  'Flavanoids',
                  'Nonflavanoid phenols',
                  'Proanthocyanins',
                  'Color intensity',
                  'Hue',
                  'OD280/OD315 of diluted wines',
                  'Proline']
#uci_wine = pd.read_csv('/kaggle/input/uci-wine/wine.data', header=None, index_col=False, names=uci_wine_names)
uci_wine = pd.read_csv('/kaggle/input/uci-wine/wine.data', header=None, names=uci_wine_names)
# Show the first instances
uci_wine.head()
```

```python
# Select the target column
uci_wine_targets = uci_wine['Target']

# Select the inputs by dropping the target column
uci_wine_inputs = uci_wine.drop('Target', axis=1)

# Show the first instances of the inputs
uci_wine_inputs.head()
```

```python
# Show simple statistical information about the dataset
uci_wine_inputs.describe()
```

```python
# Mean normalization of the inputs
uci_wine_inputs_normalized = (uci_wine_inputs-uci_wine_inputs.mean())/uci_wine_inputs.std()
# Show first instances
uci_wine_inputs_normalized.head()
```

```python
# Min max normalization
uci_wine_inputs_minmax = (uci_wine_inputs - uci_wine_inputs.min())/(uci_wine_inputs.max()-uci_wine_inputs.min())
uci_wine_inputs_minmax.head()
```

```python
# Show simple stats for mean normalized inputs
uci_wine_inputs_normalized.describe()
```

```python
# Show simple stats for min max normalized inputs
uci_wine_inputs_minmax.describe()
```

```python
# Find the percentage of instances in each class
# Through dataframe
uci_wine.groupby('Target').count()/len(uci_wine_targets)
# Through series
uci_wine_targets.value_counts()/len(uci_wine_targets)
```

```python
# Shows how to randomly split into 60/20/20

random_train = uci_wine.sample(frac=0.6)
random_val_test = uci_wine.drop(random_train.index)
random_val = random_val_test.sample(frac=0.5)
random_test = random_val_test.drop(random_val.index)

random_test['Target'].value_counts()/len(random_test)
```

```python
# Shows how to randomly split using stratified sampling

stratified_train = uci_wine.groupby('Target', group_keys=False).apply(lambda x: x.sample(frac=0.6))
stratified_val_test = uci_wine.drop(stratified_train.index)
stratified_val = stratified_val_test.groupby('Target', group_keys=False).apply(lambda x: x.sample(frac=0.5))
stratified_test = stratified_val_test.drop(stratified_val.index)

stratified_test['Target'].value_counts()/len(random_test)
```
